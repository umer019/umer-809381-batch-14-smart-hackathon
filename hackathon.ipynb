```python

last_score = 0

choice = ""

while choice != "4":
    
    print("\n===== QUIZ MENU =====")
    print("1. Start Quiz")
    print("2. View Last Score")
    print("3. Performance Analysis")
    print("4. Exit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        
        score = 0
        
        questions = [
            ["What is the capital of Pakistan?", "A) Karachi", "B) Lahore", "C) Islamabad", "D) Quetta", "C"],
            ["5 + 3 = ?", "A) 6", "B) 7", "C) 8", "D) 9", "C"],
            ["Python is a ?", "A) Animal", "B) Language", "C) Game", "D) Car", "B"]
        ]
        
        i = 0
        
        while i < len(questions):
            
            print("\n" + questions[i][0])
            print(questions[i][1])
            print(questions[i][2])
            print(questions[i][3])
            print(questions[i][4])
            
            ans = input("Enter your answer (A/B/C/D): ")
            
            while ans != "A" and ans != "B" and ans != "C" and ans != "D":
                ans = input("Invalid! Enter A/B/C/D: ")
            
            if ans == questions[i][5]:
                print("Correct ")
                score = score + 1
            else:
                print("Wrong ")
                score = score - 1   
            
            i = i + 1
        
        last_score = score
        print("\nYour Score:", score)
    
    elif choice == "2":
        print("\nLast Score:", last_score)
    
    elif choice == "3":
        
        print("\n--- Performance ---")
        
        if last_score <= 0:
            print("Level: Beginner")
            print("Feedback: Work harder!")
        elif last_score <= 2:
            print("Level: Intermediate")
            print("Feedback: Not bad ")
        else:
            print("Level: Advanced")
            print("Feedback: Excellent ")
    
    elif choice == "4":
        print("Goodbye ")
    
    else:
        print("Invalid choice! Try again.")
```
